```
import os
cd = os.getcwd()
for dirname, _, filenames in os.walk(cd):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```
import numpy as np
import pandas as pd
trainData = pd.read_csv(cd + '/train.csv')
```

```
trainData.head(5)
```

```
corrDf = trainData.corr()
corrDf
```

```
std = 0.3
features = list(corrDf[(abs(corrDf.Survived) > std) & (abs(corrDf.Survived) != 1)].index)
features
```

```
label = ['Survived']
```

```
trainDataFeatures = trainData[features]
trainDataFeatures
```

```
trainDataLabel = trainData[label]
trainDataLabel
```

```
from sklearn import linear_model
linearModel = linear_model.LinearRegression()
model = linearModel.fit(trainDataFeatures, trainDataLabel)
```

```
testData = pd.read_csv(cd + '/test.csv')
testData.head(5)
```

```
testDataFeatures = testData[features]
```

```
predict = model.predict(testDataFeatures)
predict
```

```
predictData = pd.DataFrame(predict)
predictData.columns = ['Survived']
predictData
```

```
resultData = pd.concat([testData['PassengerId'], predictData], axis=1)
resultData
```

```
resultData['Survived'] = np.where(resultData.Survived >= 0.5, 1, 0)
resultData
```

```
resultData.to_csv('gender_submission.csv', index=False)
```

```
from sklearn import ensemble
ensembleModel = ensemble.RandomForestRegressor(random_state=1, n_estimators=5)
model = ensembleModel.fit(trainDataFeatures, trainDataLabel)
```
